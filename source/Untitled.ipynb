Libraries

```python
import scipy 
import numpy as np
import matplotlib.pyplot as plt
```

Data

```python
x= np.linspace(0,2*np.pi)
y= np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# explicity
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(aspect=1)
ax.plot(x,y) 
plt.show()
```

```python
#
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(aspect=1)
ax.plot(x,y,".r",label="sin") 
ax.plot(x,y*y,".g",label="$sin^2$") 
ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
#save
plt.savefig("output.png")
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(aspect=1)
ax.plot(x,y,".",label="sin") 
ax.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.text(0.3,  0.1, "-> Mot",family="cursive",size=14)
plt.text(0.3, -0.5, "-> Mot",family="serif",size = 14)

plt.show()
```
